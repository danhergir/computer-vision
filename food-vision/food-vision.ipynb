### Importing data

Food101 dataset has 101 different classes of food and 1000 images per class.

The whole point is to speed up.

```python
# Food-101 Data Set is going to be used for this project

import requests
import zipfile
from pathlib import Path

# Setup path to a data folder
data_path = Path('data/')
image_path = data_path/'pizza_steak_sushi'

# If the image folder doesn't exist, download it
if image_path.is_dir():
    print('Already exists')

else:
    print('Creating one...')
    image_path.mkdir(parents=True, exist_ok=True)

with open(data_path / 'pizza_steak_sushi.zip', 'wb') as f:
    request = requests.get('https://github.com/mrdbourke/pytorch-deep-learning/blob/main/data/pizza_steak_sushi.zip')
    f.write(request.content)
```

### Becoming one witht the data

```python
import os

def walk_through_dir(dir_path):
    """Walks through dir_path returning its contents."""
    for dirpath, dirnames, filenames in os.walk(dir_path):
        print(f'There are {len(dirnames)} directories and {len(filenames)} images in {dirpath}')
```
